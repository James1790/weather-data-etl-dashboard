```python
import pandas as pd
import sqlite3

conn = sqlite3.connect('weather.db')
df = pd.read_sql_query("SELECT * FROM weather", conn)
```

```python
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.describe()
```

```python
df = df.drop_duplicates()
df = df.dropna(subset=['temp'])
```

```python
df.info()
```

```python
df = df.rename(columns={'temp': 'Temperature'})
df = df.rename(columns={'precip': 'Precipitation'})
df = df.rename(columns={'cond': 'Conditions'})
df = df.rename(columns={'temp_min': 'Minimum'})
df = df.rename(columns={'temp_max': 'Maximum'})
df = df.rename(columns={'city': 'City'})
df = df.rename(columns={'date': 'Date'})
df.head()
```

```python
df['City'] = df['City'].str.split(',').str[0].str.strip()
df.head()
```

```python
df.tail()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df.info()
```

```python
df.head()
```

```python
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')
```

```python
df['temp_range'] = df['maximum']-df['minimum']
df['year'] = df['date'].dt.year
df['month'] = df['date'].dt.month
df.head()
```

```python
df['conditions'] = df['conditions'].str.lower().str.strip()
df['is_rain'] = df['conditions'].str.contains('rain', na=False)
df.head()
```

```python
df = df.sort_values(by=['city','date'], ascending=[True,True])
df.head()
```

```python
desired_order = [
    'id', 'city', 'date', 'year', 'month',
    'temperature', 'minimum', 'maximum', 'temp_range',
    'precipitation', 'is_rain', 'conditions'
]

df = df[[col for col in desired_order if col in df.columns]]
```
